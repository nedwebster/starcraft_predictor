# Starcraft Predictor API

```python
import sys
sys.path.append("..")

import starcraft_predictor as scp
```

The main user API is the `scp.api.evaluate_replay()` method, which requires the user to feed a `.SC2Replay` file path, and it returns a win probability plot for that specific match.

```python
scp.api.evaluate_replay(path="../example_data/example_replay.SC2Replay")
```

The user can pass the argument `moment=True` to include the `moment` feature to the plot. This moment indicates a key change in win probability during the match, and identifies which feature was most responsible for the swing in probablity. Nn-depth documentation on the calculation of the `moment` can be found in the following notebook: `starcraft_predictor/ml_analysis/analysis_moments.ipynb`

```python
scp.api.evaluate_replay(
    path="../example_data/example_replay.SC2Replay",
    moment=True,
)
```
