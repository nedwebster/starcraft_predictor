## Example Win Probability Plot with PlotEngine

### Setup

```python
import sys
sys.path.append("..")
```

```python
import pandas as pd
import numpy as np
from numpy.random import rand

import starcraft_predictor as scp
```

### Generating Sample Data

The win probability data will arrive to the `win_probability_plot()` method as a `pandas.DataFrame` with the columns:
* `seconds` - measuring the game time in seconds,
* `win_prob` - the win probability output from the models.

We can create some sample data of this form:

```python
sample_df = pd.DataFrame(
    data={
        "seconds": np.arange(start=0, stop=600, step=10),
        "win_prob": np.clip(
            np.arange(start=0.3, stop=0.6, step=0.005) + (rand(60)/20),
            a_min=0,
            a_max=1
        )
    }
)
```

### Create Plot

A `plotnine.ggplot` object can be created from the `PlotEngine` using the `win_probability_plot()` method. We provide the other necessary arguments to the method, which will ultimately be provided on the back end by the replay itself.

```python
pe = scp.PlotEngine()
p = pe.win_probability_plot(
    df=sample_df,
    p1_race="t",
    p2_race="z",
    match_id="111111",
    p1_handle="i_play_a_skill_race",
    p2_handle="i_play_a_noob_race"
)
print(p)
```
