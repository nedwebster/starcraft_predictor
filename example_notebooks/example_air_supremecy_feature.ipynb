```python
# We can start by creating the structure of our result
# It should be a count of air supply and
# Get the first time stamp the unit is alive
def round_down_tens(n: int):
    if n % 10 == 0:
        output = n
    else:
        output = ((n + 10)  // 10) * 10
    return output

# Tests
def test_basic_usage():
    n = 105
    assert round_down_tens(n) == 110

def test_multiple_of_ten():
    n = 100
    assert round_down_tens(n) == 100

test_basic_usage()
test_multiple_of_ten()
```

```python
PROTOSS_AIR_UNITS = [
    # Omitted observer
    "Phoenix",
    "Warp Prism",
    "Void Ray",
    "Oracle",
    "Carrier",
    "Tempest",
    "Mothership",
]

PROTOSS_ANTI_AIR_UNITS = [
    "Stalker",
    "Archon",
    "Phoenix",
    "Void Ray",
    "Carrier",
    "Tempest",
]

TERRAN_AIR_UNITS = [
    "Viking",
    "Medivac",  # should we include medivacs? given the point of this is to get at air units taht can kill you since you can't kill them
    "Liberator",
    "Raven",
    "Banshee",
    "Battlecruiser",
    ]

TERRAN_ANTI_AIR_UNITS = [
    "Viking",
    "Liberator",
    "Battlecruiser",
    "Marine",
    "Ghost",  # do ghosts shoot air? include anyway for snipe?
    "Cyclone",
    "widow Mine",
    "Thor",
    "Raven",
    ]

ZERG_AIR_UNITS = [
    "Mutalisk",
    "Corruptor",
    "Brood Lord",
    "Viper",
    ]

ZERG_ANTI_AIR_UNITS = [
    "Queen",
    "Hydralisk",
    "Mutalisk",
    "Corrupter"
    ]  # Should we include: ravager, infestor?


AIR_UNITS = PROTOSS_AIR_UNITS + TERRAN_AIR_UNITS + ZERG_AIR_UNITS
ANTI_AIR_UNITS = (
    PROTOSS_ANTI_AIR_UNITS + TERRAN_ANTI_AIR_UNITS + ZERG_ANTI_AIR_UNITS
)
```

```python
def get_unit_living_range(unit: Unit, max_died_at: int) -> tuple:
    unit_created = round_down_tens(unit.finished_at / 16)
    unit_died = round_down_tens(max_died_at) if unit.died_at is None else round_down_tens(unit.died_at / 16)
    
    return unit_created, unit_died
```

```python
# Initialize a pandas frame with air and anti air supply fields
data = pd.DataFrame({
    "seconds": range(0, replay.events[-1].second, 10),
    "player_1_air_supply": 0,
    "player_2_air_supply": 0,
    "player_1_anti_air_supply": 0,
    "player_2_anti_air_supply": 0,  
})
```

```python
for unit in units:
    if unit.owner is None:
        next
    unit_created, unit_died = get_unit_living_range(
        unit=unit, max_died_at=replay.events[-1].second + 10
    )
    if unit.name in AIR_UNITS:
        data.loc[
            data.seconds.between(unit_created, unit_died),
            f"player_{unit.owner.team_id}_air_supply"
        ] = data.loc[
            data.seconds.between(unit_created, unit_died),
            f"player_{unit.owner.team_id}_air_supply"
        ] + unit.supply
    if unit.name in ANTI_AIR_UNITS:
        data.loc[
            data.seconds.between(unit_created, unit_died),
            f"player_{unit.owner.team_id}_anti_air_supply"
        ] = data.loc[
            data.seconds.between(unit_created, unit_died),
            f"player_{unit.owner.team_id}_anti_air_supply"
        ] + unit.supply
```
